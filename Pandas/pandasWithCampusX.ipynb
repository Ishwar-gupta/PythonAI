```python
def get_matches_count(city):
    mask=data['city']==city
    return data[mask].shape[0]
get_matches_count('Bangalore') # 66
get_matches_count('Rajkot') # 10
```

### to know which team got winner and how many times

### the pandas plot() method

```python
import matplotlib.pyplot as plt 
data['winner'].value_counts().plot(kind='bar')
plt.ylabel("number of match win")
plt.title("team got winner")
plt.grid(True)

```

### making pie-chart of toss_decision how many team chose  fielding and bating

```python
data['toss_decision'].value_counts().plot(kind='pie')
```

### drawing histogram of win_by_runs with it's frequency

```python
data['win_by_runs'].plot(kind='hist')
plt.title('win_by_runs')
```

### Calculating total matches played by team as in both team1 and team2

```python
data['team1'].value_counts() + data['team2'].value_counts()
```

### sorting total matches in ascending order

```python
(data['team1'].value_counts() + data['team2'].value_counts()).sort_values()
```

### sorting all matches in descending order

```python
(data['team1'].value_counts() + data['team2'].value_counts()).sort_values(ascending=False)
```

### sorting dataframe based on city names
