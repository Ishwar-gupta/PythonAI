```python
# 2.Drop rows with missing data using dropna().
import pandas as pd

data = pd.DataFrame({
    "Name": ["Ram", "Sita", "Hari", "Gita"],
    "Marks": [85, None, 90, None],
    "City": ["Butwal", "Pokhara", None, "Kathmandu"]
})

print("Original DataFrame:")
print(data)
```

```python
# using axis=1 for specific columns if there is any missing value(NaN)
# drop columns instead of rows:
data3=data.dropna(axis=1)
data3
```

```python
# Drop rows where all values NaN using how='all'
# i.e it'll only remove if all values are NaN
data4=data.dropna(how='all')
data4
```

### üîπ 5. Statistical & Aggregation Functions

1.Find the average, median, and standard deviation of Marks.

2.Display the student with the highest Marks.

3.Count how many students are from each City using value_counts().

4.Use agg() to calculate both mean and max of Marks.

```python
# 1. finding the avg,Md & Std of marks
avg=dat['Marks'].mean()
median=dat['Marks'].median()
std_dev=dat['Marks'].std()
print("Average of the marks:",avg)
print("Median of the marks:",median)
print("Standard Deviation of the marks:",std_dev)
```

```python
# 2.Displaying the student with highest marks using .loc() and idxmax() method
top_std=dat.loc[dat['Marks'].idxmax()]
print("Student's with highest marks:\n",top_std)
last_std=dat.loc[dat['Marks'].idxmin()]
print("Student's with lowest marks:\n",last_std)
```

```python
# Counting how many students are from each city
city_count=dat['City'].value_counts()
print("Number of students from each city:\n",city_count)
```

```python
# Using agg() method to calculate both mean ,median,std,min and max of the marks
aggregate=dat['Marks'].agg(['mean','max','std','min','median'])
print("Aggregation Results:\n",aggregate)
```

### üîπ 6. Grouping & Aggregation

1. Group the data by City and find the average Marks for each city.

2. Group by Gender and find the number of students and average marks.

3. Display the top student (highest marks) from each city using groupby() + idxmax().

```python

dgp = pd.DataFrame({
    "Name": ["Ram", "Sita", "Hari", "Gita", "Nabin", "Rita", "Kiran"],
    "Marks": [80, 90, 85, 70, 95, 88, 76],
    "City": ["Butwal", "Pokhara", "Parasi", "Kathmandu", "Dharan", "Pokhara", "Butwal"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M"]
})
# 1.) grouping the data by city and average marks for each city
avg_marks_by_city=dgp.groupby("City")['Marks'].mean()
avg_marks_by_city
```

```python
# 2.) grouping by gender & finding the no. of std and average
genderStatus=dgp.groupby('Gender')['Marks'].agg(['count','mean'])
genderStatus
```

```python
# finding the top student using groupby() and idxmax() method
top_student=dgp.loc[dgp.groupby('City')['Marks'].idxmax()]
print("Top Student from each Student:\n",top_student)
```

### üîπ 11. Visualization (optional but great practice)

1. Plot a bar chart of average marks per city using groupby() + .plot.bar().

2. Create a histogram of marks distribution using .plot.hist().

```python
import matplotlib.pyplot as plt
#1.) Group by City and find average marks
avg_marks = dgp.groupby('City')['Marks'].mean()

# Plot a bar chart
avg_marks.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Marks per City')
plt.xlabel('City')
plt.ylabel('Average Marks')
plt.show()
```
