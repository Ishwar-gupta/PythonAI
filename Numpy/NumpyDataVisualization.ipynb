```python
import numpy as np
import matplotlib.pyplot as plt

# Data structure:[Resturant Id: 2021 , 2022 , 2023 , 2024]
sales_data=np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   #Chai Point
])
print("======= Zomato sales analysis  =========")
print("\n Sales data shape:",sales_data.shape)
print("\n*******    Sample data from first to 3rd restaurant  ********")
print("\n Sample data from first row to third row:\n\n",sales_data[:3]) # [0:3]==[:3]
# =======    Total sales per year  ==========
print("\n =======    Total sales per year  ==========\n")
print("total sales per year:",np.sum(sales_data,axis=0))
print("total sales per year without ID:",np.sum(sales_data[:, 1:],axis=0))
# ************** Minmum sales per restaurant**********************************
print("\n******* Minmum sales per restaurant and per year **********\n")
min_sales=np.min(sales_data[:, 1:],axis=1)
print("Minimum Sales per restaurant:",min_sales)
min_sales_per_year=np.min(sales_data[:, 1:],axis=0)
print("Minimum Sales per year:",min_sales_per_year)
# ************** Maximum sales per restaurant**********************************
print("\n******* Maximum sales per restaurant and per year **********\n")
max_sales=np.max(sales_data[:, 1:],axis=1)
print("Maximum Sales per restaurant:",max_sales)
max_sales_per_year=np.max(sales_data[:, 1:],axis=0)
print("Maximum sales per year:",max_sales_per_year)
# *************  Average Sales per year  ***********************
print("\n******* Maximum sales per restaurant and per year **********\n")
print("Average Sales per restaurant:",np.mean(sales_data[:, 1:],axis=1))
print("Average Sales per year:",np.mean(sales_data[:, 1:],axis=0))
cumsum=np.cumsum(sales_data[:, 1:],axis=1)
print("cumulative sum:",cumsum)
#   *************    matplotlib    **********************************
plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales across all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()





```
